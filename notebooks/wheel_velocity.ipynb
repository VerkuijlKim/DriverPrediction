```python
df_interesting = df[['Class','Wheel_velocity_front_left-hand',
   'Wheel_velocity_rear_left-hand', 
   'Wheel_velocity_rear_right-hand', 
   'Wheel_velocity_front_right-hand',
   'Vehicle_speed',
   'Acceleration_speed_-_Longitudinal',
   'Acceleration_speed_-_Lateral']].copy()
```

```python
fig = plt.figure(figsize=(10, 7))

# Group data and extract lists of values
data = [group['Acceleration_speed_-_Longitudinal'].values for _, group in df_interesting.groupby('Class')]

# Create boxplot
plt.boxplot(data, labels=df_interesting['Class'].unique())

# Show plot
plt.show()
```

```python
#track: front/rear = 62 inch, 62.4 inch
front_track = 0.0254*62
rear_track = 0.0254*62.4
wheelbase = 2.550
avg_track = (front_track+rear_track)/2

print(front_track,rear_track,wheelbase)
```

```python
import numpy as np

# Create an empty list to store results
g_forces = []

for _, row in df_interesting.iterrows():
    # Extract wheel velocities from the row
    V_FL = row['Wheel_velocity_front_left-hand']
    V_FR = row['Wheel_velocity_front_right-hand']
    V_RL = row['Wheel_velocity_rear_left-hand']
    V_RR = row['Wheel_velocity_rear_right-hand']

    # Compute Yaw Rate
    omega = ((V_FR + V_RR) - (V_FL + V_RL)) / (2 * avg_track)

    # Compute Vehicle Speed
    V = (V_FL + V_FR + V_RL + V_RR) / 4

    # Compute Turning Radius
    R = V / omega if omega != 0 else float('inf')  # Avoid division by zero

    # Compute Lateral Acceleration (m/sÂ²)
    a_y = V**2 / R if R != float('inf') else 0

    # Convert to g-forces
    g_lat = a_y / 9.81

    # Store result
    g_forces.append(g_lat)

# Add computed g-forces as a new column in the DataFrame
df_interesting['g_force_lateral'] = g_forces
```
